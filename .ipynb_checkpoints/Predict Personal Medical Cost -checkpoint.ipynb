```python
import pandas as pd
import seaborn as sns
```

```python
sns.set_palette("Accent")
sns.set_style("darkgrid")
```

```python
df = pd.read_csv('./data/insurance.csv')
df
```

```python
df.describe()
```

```python
df.groupby('region')['region'].count()
```

```python
df.groupby('smoker')['smoker'].count()
```

```python
def lambda_region(x):
    if x == 'northeast':
        return 0
    if x == 'northwest':
        return 1 
    if x == 'southeast':
        return 2
    else:
        return 3
```

```python
df['binario_smoker'] = df['smoker'].apply(lambda x: 1 if x == 'yes' else 0)
df['binario_sex'] = df['sex'].apply(lambda x: 1 if x == 'female' else 0)
df['binario_region'] = df['region'].apply(lambda_region)
df.head()
```

```python
df['log_age'] = np.log(df['age'])
df['log_bmi'] = np.log(df['bmi'])
df['log_charges'] = np.log(df['charges'])
df['log_binario_smoker'] =  np.log(df['binario_smoker'] + 1)
df['log_region'] = np.log(df['binario_region'] + 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['log_charges']
```

```python
X = df[['log_age', 'log_bmi', 'log_binario_smoker']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=2811)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ml1 = LinearRegression()
```

```python
ml1.fit(X_train, y_train)
```

```python
ml1.score(X_test, y_test)
```
